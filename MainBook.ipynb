```python
import pandas as pd
categories_frame = pd.read_csv(filepath_or_buffer="./categories.csv",sep=";")
categories_frame
```

```python
sales_frame = pd.read_csv(filepath_or_buffer="./sales.csv",sep=";" ) #,nrows=1000000)
sales_frame
```

```python
products_frame = pd.read_csv(filepath_or_buffer="./products.csv",sep=";")
products_frame
```

```python
import matplotlib.pyplot as plt
```

```python
merge_frame = sales_frame.join(products_frame.set_index('ProductID'), on='ProductID')
merge_frame=merge_frame.join(categories_frame.set_index('CategoryID'), on='CategoryID')
merge_frame[0:1000].to_csv(path_or_buf="merge products sales categories.csv", sep=';')
```

```python
import seaborn as sns
categoryquantity = merge_frame[['CategoryName','Quantity'] ].groupby(['CategoryName']).sum() 

categoryquantity.reset_index(inplace=True)
a4_dims = (11.7, 8.27)
plt.subplots(figsize=a4_dims)
sns_plot = sns.barplot(x='CategoryName', y='Quantity', data=categoryquantity);
sns_plot.get_figure().savefig("All sales in quantity per category.png")
```
